Handling Imbalanced Data

```python
def ANN(x_train,y_train,x_test,y_test,loss,weights):
  model = keras.Sequential([
                          keras.layers.Dense(64,input_shape=(x_train.shape[1],),activation="relu"),
                          keras.layers.Dense(32,activation="relu"),
                          keras.layers.Dense(1,activation="sigmoid")
  ])

  model.compile(optimizer="adam",loss=loss,metrics=["accuracy"])

  if weights == -1 :
    model.fit(x_train,y_train,epochs=100)
  else:
    model.fit(x_train,y_train,epochs=100,class_weight=weights)

  print(model.evaluate(x_test,y_test))

  yp = model.predict(x_test)
  yp = np.round(yp)

  print("Classification Report: \n",classification_report(y_test,yp))

  return yp
```

Under Sampling

```python
class_count_0,class_count_1 = df.Class.value_counts()
```

```python
df_class_0 = df[df["Class"] == 0]
df_class_1 = df[df["Class"] == 1]
```

```python
df_class_0.shape , df_class_1.shape
```

```python
df_class_0_under = df_class_0.sample(class_count_1)
```

```python
pd.concat([df_class_0_under,df_class_1]).Class.value_counts()
```

```python
df_under = pd.concat([df_class_0_under,df_class_1])
```

```python
df_under
```

```python
x= df_under.drop("Class",axis=1)
y=df_under["Class"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(x_train,y_train,x_test,y_test,"binary_crossentropy",-1)
```

```python
def ANN2(x_train,y_train,x_test,y_test,loss,weights):
  model = keras.Sequential([
                          keras.layers.Dense(64,input_shape=(x_train.shape[1],),activation="relu"),
                          keras.layers.Dense(32,activation="relu"),
                          keras.layers.Dense(1,activation="sigmoid")
  ])

  model.compile(optimizer="adam",loss=loss,metrics=["accuracy"])

  if weights == -1 :
    model.fit(x_train,y_train,epochs=3)
  else:
    model.fit(x_train,y_train,epochs=3,class_weight=weights)

  print(model.evaluate(x_test,y_test))

  yp = model.predict(x_test)
  yp = np.round(yp)

  print("Classification Report: \n",classification_report(y_test,yp))

  return yp
```

```python
y_preds = ANN2(x_train,y_train,x_test,y_test,"binary_crossentropy",-1)
```
